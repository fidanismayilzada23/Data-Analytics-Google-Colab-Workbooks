```python
data={
    'year':[2019,2020,2021,2022,2023,2024,2025],
    'population':[9981500,10023000,10027000,10050000,10100000,10180770,10423252]
}
```

```python
import pandas as pd
df=pd.DataFrame(data)
plt.figure(figsize=(10,6))
plt.gca().set_facecolor('#fcccf6')  #get current axes
plt.grid(True,linestyle='--',alpha=0.5)
plt.plot(df['year'],df['population'],color='#211720',linewidth=3,marker='o',markersize=10,markerfacecolor='red')

plt.title("Population of country(2019-2025)",fontsize=16,fontweight='bold',color='darkslategray')
plt.xlabel('Year',fontsize=12,fontweight='bold',color='darkslategray')
plt.ylabel('Population',fontsize=16,fontweight='bold',color='darkslategray')

#üstündeki reqemler
for x,y in zip (df['year'],df['population']):
  plt.text(x,y,f'{y:,}',ha='right',fontsize=10,fontweight='bold')
```

#London Bike Analis

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("hmavrodiev/london-bike-sharing-dataset")

print("Path to dataset files:", path)
```

```python
df=pd.read_csv(path + '/london_merged.csv')
df.head()
```

```python
#data time object dipinde olmamalıdır / yanlış var data time çevrilmelidir
df.info()
```

```python
df.isnull().sum()  #temiz datadır
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
df['timestamp']=pd.to_datetime(df['timestamp'])
df.dtypes
#time date time çevirdik
```

```python
df['hour']=df['timestamp'].dt.hour
df['month']=df['timestamp'].dt.month
df['day']=df['timestamp'].dt.day
df.head()
```

```python
df['is_weekend'].value_counts()
```

```python
h=df.groupby(df['is_weekend'])['cnt'].mean()
plt.bar(['weekday','weekend'],h.values)
#hefteiçi daha çoxdur satış
```

```python
df.groupby(df['hour'])['cnt'].mean().plot(kind='bar')
```

```python
df.groupby(df['month'])['cnt'].mean().plot(kind='bar')
```

```python
df['season'].unique()
```

```python
season_map={0:'spring',
            1:'summer',
            2:'fall',
            3:'winter'
             }
s=df.groupby(df['season'])['cnt'].mean()
plt.pie(s.values,labels=s.index.map(season_map),autopct='%.1f%%');
```

```python
df.corr()
```

```python
plt.scatter(df['t1'],df['cnt']);
```

```python
h=df.groupby(df['hour'])['cnt'].mean()
plt.figure(figsize=(10,7))
plt.plot(h.index,h.values)
plt.title('Hour vs Cnt')
plt.xlabel('Hour')
plt.ylabel('Cnt')
plt.grid()
plt.show()
```
